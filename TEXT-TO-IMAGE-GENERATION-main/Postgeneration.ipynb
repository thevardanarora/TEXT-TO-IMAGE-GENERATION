```python
import torch
import gradio as gr
from diffusers import StableDiffusionXLPipeline

# Load the SSD-1B model from Segmind using the pre-trained pipeline
pipe = StableDiffusionXLPipeline.from_pretrained(
    "segmind/SSD-1B",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)

# Move model to GPU (if available)
pipe.to("cuda")

# Optional: Enable memory-efficient attention if you're using torch < 2.0
# pipe.enable_xformers_memory_efficient_attention()

# Function to generate the image
def generate_image(prompt, neg_prompt="ugly, blurry, poor quality"):
    # Generate the image using the provided prompt
    image = pipe(prompt=prompt, negative_prompt=neg_prompt).images[0]
    return image

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Segmind SSD-1B Image Generator")

    prompt_input = gr.Textbox(label="Prompt", placeholder="Enter a description for the image...")
    neg_prompt_input = gr.Textbox(label="Negative Prompt", value="ugly, blurry, poor quality")

    image_output = gr.Image(label="Generated Image")

    submit_button = gr.Button("Generate Image")

    # Connect the function to the submit button
    submit_button.click(
        fn=generate_image,
        inputs=[prompt_input, neg_prompt_input],
        outputs=image_output
    )

# Launch the Gradio interface
demo.launch()
```
